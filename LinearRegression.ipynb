```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/insurance.csv')
```

```python
df.info()
```

```python
x=df['age'].values
y=df['charges'].values
xmean=np.mean(x)
ymean=np.mean(y)
n=len(x)
nume=0
denom=0
for i in range(n):
  nume+=(x[i]-xmean)*(y[i]-ymean)
  denom+=(x[i]-xmean)**2
m=nume/denom
c= ymean - m*xmean
print(m,'\n',c,'\n')
X = np.linspace(18,65,1000)
Y=c+m*X

plt.plot(X, Y, color='#00ff00', label='Linear Regression')
plt.scatter(x, y,label='Linear Regression')
plt.show()
```

```python
x=df['age'].values
y=df['charges'].values
xmean=np.mean(x)
ymean=np.mean(y)
n=len(x)
nume=0
denom=0
for i in range(n-1):
  nume+=(x[i+1]-x[i])
  denom+=y[i+1]-y[i]
m=nume/denom
c= ymean - m*xmean
print(m,'\n',c,'\n')
X = np.linspace(18,65,1000)
Y=c+m*X

plt.plot(X, Y, color='#00ff00', label='Linear Regression')
plt.scatter(x, y,label='Linear Regression')
plt.show()
```
